```python
import pandas as pd

df = pd.read_csv("Mall_Customers.csv")
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder, StandardScaler

df['Gender'] = LabelEncoder().fit_transform(df['Gender'])  # Male=1, Female=0
X = df[['Age', 'Annual Income (k$)', 'Spending Score (1-100)']]  # or include Gender
X_scaled = StandardScaler().fit_transform(X)
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

wcss = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, init='k-means++', random_state=42)
    kmeans.fit(X_scaled)
    wcss.append(kmeans.inertia_)

plt.plot(range(1, 11), wcss)
plt.xlabel("Number of Clusters")
plt.ylabel("WCSS")
plt.title("Elbow Method")
plt.show()
```

```python
kmeans = KMeans(n_clusters=5, init='k-means++', random_state=42)
clusters = kmeans.fit_predict(X_scaled)
df['Cluster'] = clusters
```

```python
import seaborn as sns

sns.scatterplot(x='Annual Income (k$)', y='Spending Score (1-100)', hue='Cluster', data=df, palette='Set1')
plt.title("Customer Segments")
plt.show()
```

```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Sample dataset
data = pd.DataFrame({
    'Age': [19, 35, 26, 27, 19],
    'Annual Income (k$)': [15, 35, 75, 60, 40],
    'Spending Score (1-100)': [39, 81, 6, 77, 40]
})

# Step 1: Scale the data
scaler = StandardScaler()
scaled_data = scaler.fit_transform(data)

# Step 2: Train the KMeans model
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(scaled_data)

# Step 3: New data to predict
new_data = pd.DataFrame([[30, 60, 75]], columns=['Age', 'Annual Income (k$)', 'Spending Score (1-100)'])

# Step 4: Scale the new data using the same scaler
new_data_scaled = scaler.transform(new_data)

# Step 5: Predict the cluster
cluster_label = kmeans.predict(new_data_scaled)
print("Predicted cluster:", cluster_label[0])
```

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.cluster import KMeans

# Sample customer dataset
data = pd.DataFrame({
    'Age': [18, 22, 25, 30, 35, 40, 45, 50, 60, 65],
    'Annual Income (k$)': [15, 18, 20, 25, 30, 35, 40, 50, 60, 70],
    'Spending Score (1-100)': [39, 81, 6, 77, 40, 60, 20, 80, 30, 50]
})

# Step 1: Scale the original data
scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(data)

# Step 2: Train KMeans model with 5 clusters
kmeans = KMeans(n_clusters=5, random_state=0)
kmeans.fit(scaled_data)

# Step 3: New customer data
new_data = pd.DataFrame([[55, 45, 90]], columns=['Age', 'Annual Income (k$)', 'Spending Score (1-100)'])

# Step 4: Scale new data using the same scaler
new_data_scaled = scaler.transform(new_data)

# Step 5: Predict cluster for new data
cluster_label = kmeans.predict(new_data_scaled)

print("Predicted Cluster for new customer:", cluster_label[0])
```
